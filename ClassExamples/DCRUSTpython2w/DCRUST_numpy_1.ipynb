```
a2 = np.array(
    [ 
     [2,3,4,5] ,
     [6,7,8,9] 
    ])
```

```
print(a2.shape)
print(a2.size)
print(a2.ndim)
print(a2.dtype)
```

```
print( np.ones((2,3)) )
print( np.zeros((2,3,2)) )
```

```
print(a2)
print("\na2[0,0]   ", a2[0,0])
print("a2[1,2]   ", a2[1,2])
print("a2[1]     ", a2[1])
print("a2[0,1:3] ", a2[0,1:3]) #slice
print("a2[0:2,1:3] ", a2[0:2,1:3])
```

```
a3 = np.arange(8)
print(a3)
print(a3[1:5])
print(a3[:5])
print(a3[3:])
print(a3[:])
```

```
# change the data
print(a3)
a3[4] = 90
print(a3)
```

```

a3 = np.arange(8)
print(a3)
a3[2:5] = 9
print(a3)
a3[2:5] = [9,99,199]
print(a3)
#a3[2:5] = [9,99,199,299,399,499]
#a3[2:5] = [9,99 ]
a3[2:5] = [9,99,10.10]
print(a3)
```

```
a4 = np.arange(12)
print(a4)
a4 = a4.reshape(3,4)
print(a4)
print("\na4[0:2 , 1:3]  ", a4[0:2 , 1:3])
print("\na4[: , 1:3]  ", a4[: , 1:3])  
```

```
a4 = np.arange(24)
print(a4)
a4 = a4.reshape(3,4,2)
print(a4)
print (a4[:, 1:3 , 1])
# for non contigous
print (a4[:, [1,3] , 1])
print (a4[:,: , [1]])
print (a4[..., 1])
```
