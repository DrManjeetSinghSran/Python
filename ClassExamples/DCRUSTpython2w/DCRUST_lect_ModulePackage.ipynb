```
def areaTriangle(base, height):
  return .5 * base * height
```

```
areaTriangle(5,4)
```

```
def areaSquare(side):
  return side * side
```

```
areaSquare(10)
```

```
def perimeterSquare(side):
  return 4*side
```

```
perimeterSquare(40)
```

```
def perimeterRectangle(length, bredth):
  return 2*(length + bredth)
```

```
perimeterRectangle(4,2)
```
