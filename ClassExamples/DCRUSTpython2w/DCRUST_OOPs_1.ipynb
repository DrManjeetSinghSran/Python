## Object-oriented Programming vs	Procedural Programming


1.	Object-oriented programming is the problem-solving approach and used where computation is done by using objects.	Procedural programming uses a list of instructions to do computation step by step.
2.	It makes the development and maintenance easier.	In procedural programming, It is not easy to maintain the codes when the project becomes lengthy.
3.	It simulates the real world entity. So real-world problems can be easily solved through oops.	It doesn't simulate the real world. Procedural works on step by step instructions divided into small parts called functions.
4.	It provides data hiding. So it is more secure than procedural languages. You cannot access private data from anywhere.	Procedural language doesn't provide any proper way for data binding, so it is less secure.
5.	Example of object-oriented programming languages is C++, Java, .Net, Python, C#, etc.	Example of procedural languages are: C, Fortran, Pascal, VB etc.

To make fee variable and its validation logic work together we put them in the class

* Class − A user-defined prototype for an object that defines a set of attributes that characterize any object of the class. The attributes are data members (class variables and instance variables) and methods, accessed via dot notation.
* Class variable − A variable that is shared by all instances of a class. Class variables are defined within a class but outside any of the class's methods. Class variables are not used as frequently as instance variables are.
* Data member − A class variable or instance variable that holds data associated with a class and its objects.
* Function overloading − The assignment of more than one behavior to a particular function. The operation performed varies by the types of objects or arguments involved.
* Instance variable − A variable that is defined inside a method and belongs only to the current instance of a class.

## Constructor

* having name as  _ _init_ _
* it is a method which is called automatically/implicitly
* user cannot call it explicitly


```
class Student:
  count=0
  def __init__(self):
    #print("calling constructor __init__ ")
    self.fee=10
    Student.count += 1   #increment the value of count

#the variable having self as prefix , they are local/object variable
# count is class variable , and accessed with class name 
```

```
stu1 = Student()
print(stu1)
print("stu1.count     : ",stu1.count)
print("stu1.fee       :",stu1.fee)
print("Student.count  : ", Student.count)
```

## Methods in Class

```
class Student:
  count=0
  def __init__(self ,sName='', sFee=0) : 
    self.fee  =  sFee
    self.name =  sName
    Student.count += 1 

  def showStudent(self): 
    print("Name   :",self.name)
    print("Fee    :",self.fee)
    print("Count  :" , Student.count) 
```

## Method overloading

```
def add(a, b ):
  print(a+b)
```

```
add(10,20)
```

```
def add(a, b,c ):
  print(a+b+ c)
```

```
add(10,20,30)
```

```
# best way is to use default argument to emulate overloading
def add(a, b,c=0):
  print(a+b+ c)
```

```
add(10,20)
add(10,20,30)
```

```
#overloading showStudent method
class Student:
  count=0
  def __init__(self ,sName='', sFee=0) : 
    self.fee  =  sFee
    self.name =  sName
    Student.count += 1 

  def showStudent(self, fillchar=':'): 
    print("Name   ",fillchar,self.name)
    print("Fee    ",fillchar,self.fee)
    print("Count  " ,fillchar, Student.count)

```

```
stu6 = Student('Bhagat Singh' , 909)
stu6.showStudent('* ')
```

```
stu6.showStudent(' % ')
```

```
stu6.showStudent()
```

```
print(stu6)
```

```
#overloading String method of class
class Student:
  count=0
  def __init__(self ,sName='', sFee=0) : 
    self.fee  =  sFee
    self.name =  sName
    Student.count += 1 

  def showStudent(self, fillchar=':'): 
    print("Name   ",fillchar,self.name)
    print("Fee    ",fillchar,self.fee)
    print("Count  " ,fillchar, Student.count)

  def __str__ (self ) : 
    #return 'this is Student class'
    return '{0} has fee = {1}'.format(self.name, self.fee)
```

```
stu7 = Student('Himmat Singh')
```

```
stu7.showStudent()
print(stu7)
```
