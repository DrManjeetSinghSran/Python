## Accessing attributes without class methods




* getattr(obj, name[, default]) − to access the attribute of object.
* hasattr(obj,name) − to check if an attribute exists or not.
* setattr(obj,name,value) − to set an attribute. If attribute does not exist, then it would be created.
* delattr(obj, name) − to delete an attribute.

````
class Student:
  '''this is Student class first example in OOPs ''' 

  count=0
  def __init__(self ,sName='', sFee=0) : 
    # fee, name are the attributes of this class
    self.fee  =  sFee
    self.name =  sName
    Student.count += 1 

  def showStudent(self, fillchar=':'): 
    print("Name   ",fillchar,self.name)
    print("Fee    ",fillchar,self.fee)
    print("Count  " ,fillchar, Student.count)

  def __str__ (self ) : 
    #return 'this is Student class'
    return '{0} has fee = Rs.{1}/-'.format(self.name, self.fee)
````

````
stu1 = Student('suresh kumar', 1000)
print(stu1)
````

````
print(hasattr(stu1, 'fee'))
print(hasattr(stu1, 'feepaid'))
if hasattr(stu1, 'fee') : print (stu1.fee)
````

````
print(getattr(stu1, 'fee'))
````

````
# change/set the value of attribute
print (stu1.fee)
stu1.fee=3000
print (stu1.fee)
setattr(stu1, 'fee' , 3500)
print (stu1.fee)
````

## Destroying Objects (Garbage Collection)

* Python deletes unneeded objects (built-in types or class instances) automatically to free the memory space. 
* Python's garbage collector runs during program execution and is triggered when an object's reference count reaches zero. An object's reference count changes as the number of aliases that point to it changes.
* destructor __ _del_ __()

````
class Student:
  '''this is Student class first example in OOPs ''' 

  count=0
  def __init__(self ,sName='', sFee=0) : 
    # fee, name are the attributes of this class
    self.fee  =  sFee
    self.name =  sName
    Student.count += 1 

  def showStudent(self, fillchar=':'): 
    print("Name   ",fillchar,self.name)
    print("Fee    ",fillchar,self.fee)
    print("Count  " ,fillchar, Student.count)

  def __str__ (self ) : 
    #return 'this is Student class'
    return '{0} has fee = Rs.{1}/-'.format(self.name, self.fee)

  def __del__(self) : 
    Student.count -= 1 
    print("Destructor called")
````

# Inheritance

## Type of Inheritance

Basic Syntax
```
class derived-class(base class):  
    <class-suite>
```
Multiclass (multiple) Inheritance
```
class derive-class(<base class 1>, <base class 2>, ..... <base class n>):  
    <class - suite>  
```
Multilevel Inheritance
```
class class1:  
    <class-suite>   
class class2(class1):  
    <class suite>  
class class3(class2):  
    <class suite> 
``` 

````
# basic inheritance
class A :
  def __init__ (self, var=0) :
    print("A constructor ")
    self.a= var

  def __str__ (self ) :
    return str(self.a)

  def __del__ (self) :
    print("A destructor ")

  def showA(self):
    return "a={0}".format(self.a)
````

````
a1 = A()
print(a1)
print( A(10))   # anonymous object 
````

````
# inherit class A (base) into class B (derived )
class B (A):
  def __init__ (self, var1=0, var2=0) :
    print("B constructor ")
    #call the constructor of A class (base class)
    super().__init__ (var1)
    self.b= var2

  def __str__ (self ) :
    #return str(self.b)
    return 'a={0}  b={1}'.format(self.a, self.b)

  def __del__ (self) :
    print("B destructor ")

  def showB(self):
    return "b={0}".format(self.b)
````

## Data Hiding (public,private, protected)

````
# by default all variables are public 

class A :
  
  iPub = None    # public data member
  _iPro = None   # protected data member 
  __iPri = None  # private data member 

  def __init__ (self, iPub=0,iPro=0,iPri=0):

    self.iPub = iPub
    self._iPro= iPro     #   protected attribute 
    self.__iPri= iPri     #   private attribute 

  def showA(self) :
    print ( "iPub : ",self.iPub , ", iPro : ",self._iPro, ", iPri : ", self.__iPri)
````

````
a1 = A(10,20,30)
a1.showA()
````

````
 # only public variable are available outside the class using object
 print(a1.iPub)
````

````
 print(a1._iPro)  # protect member access
````

````
print( a1._A__iPri) # private member access
a1._A__iPri += 20
print( a1._A__iPri)
````

````
# derived class
class B(A):
  def showB (self) :
    print ( "iPub : ",self.iPub , ", iPro : ",self._iPro )  #, iPri : ", self.__iPri)

  def incAll(self) : 
    self.iPub +=10
    self._iPro +=20
#derived class can only access public and protected variable
````

````
b1 = B ()
````

````
b1.showB()
b1.incAll()
b1.showB()
````

**Protected Access Modifier**

 * The members of a class that are declared protected are only accessible to a class derived from it. Data members of a class are declared protected by adding a single underscore ‘_’ symbol before the data member of that class.

**Private Access Modifier**

* The members of a class that are declared private are accessible within the class only, private access modifier is the most secure access modifier. Data members of a class are declared private by adding a double underscore ‘__’ symbol before the data member of that class.

* Every member with double underscore will be changed to _object._class__variable. If so required, it can still be accessed from outside the class, but the practice should be refrained. a1.**_class__iPri**

# Operator overloading

````
class Point:
  def __init__ (self, x=0,y=0):
    #print(" constructor fired ") 
    self.x = x
    self.y = y
  def __str__ (self) :
    return "({0},{1})".format (self.x, self.y)

  #operator overloading
  def __add__(self , A):
    return Point( self.x + A.x , self.y + A.y )
````

````
P = Point(2,3)
Q = Point(5,8)
print(P,Q)
````

````
R = P + Q
print(R)
````

````
#anonymous
print (  Point(1,1)+ Point(2,2) )
````
