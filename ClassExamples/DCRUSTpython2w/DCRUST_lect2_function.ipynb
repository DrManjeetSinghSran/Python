## Parameter passing

```
# define method to print first n natural number -> n to be sent by user
# function with parameter

def naturalNumber1toN( n ):
  '''function to print first n natural numbers '''
  #function body / logic 
  for i in range(1,n+1):
    print (i , end =" ")
```

```
#naturalNumber1toN(5)
naturalNumber1toN(15)
 
```

```
# function with multiple parameter

def naturalNumberMtoN(start, end  ):
  '''function to print first n natural numbers '''
  #function body / logic 
  for i in range(start ,end+1):
    print (i , end =" ")
```

```
naturalNumberMtoN(3,30)
```

## return value from function 

```
# function returning value 

def sumNaturalNumberMtoN(start, end  ):
  sum = 0
  for i in range(start ,end+1):
    sum += i
    #print (i , end =" ")
  return (sum)
```

```
#sumNaturalNumberMtoN(2,5)
print(sumNaturalNumberMtoN(2,5) )
print(sumNaturalNumberMtoN(2,5) +10)
print(sumNaturalNumberMtoN(2,5) + sumNaturalNumberMtoN(50,55)  )
print(sumNaturalNumberMtoN(5,10))
print( sumNaturalNumberMtoN(2,  sumNaturalNumberMtoN(5,10)  ) )
```

## call by value/reference

```
# Address of variable 
m=100
print("m : " , m , "  id(m) : ", id(m))
n=m
print("n : " , n , "  id(n) : ", id(n))
p=100
print("p : " , p , "  id(p) : ", id(p))
p+=1 ; print("p+=1")
print("m : " , m , "  id(m) : ", id(m))
print("n : " , n , "  id(n) : ", id(n))
print("p : " , p , "  id(p) : ", id(p))
p=100 ; print("p=100")
print("m : " , m , "  id(m) : ", id(m))
print("n : " , n , "  id(n) : ", id(n))
print("p : " , p , "  id(p) : ", id(p))
```

```
def changeMe(a,b): 
  print("a : " , a , "  id(a) : ", id(a) , "  b : " , b , "  id(b) : ", id(b))
  a+=1
  b-=1
  print("a : " , a , "  id(a) : ", id(a) , "  b : " , b , "  id(b) : ", id(b))
```

```
x= 10 ; y=20
print("x : " ,  x ,"  id(x) : ", id(x) , "  y : ", y ,"  id(y) : ", id(y))
changeMe(x,y)
print("x : " ,  x ,"  id(x) : ", id(x) , "  y : ", y ,"  id(y) : ", id(y))
```

```
def changeMe1(arr):
  print("\nValues received in method : ")
  for i in arr:   print(i, end=" ")
  arr[0]+=11
  print("\nAfter changing in method : ")
  for i in arr:   print(i, end=" ")
```

```
arr = [10,20,30,40]
print("\nBefore Passing in method : ")
for i in arr:   print(i, end=" ")
changeMe1(arr)
print("\nAfter returning from method : ")
for i in arr:   print(i, end=" ")
```

## Default argument 

```
# default value can be anything 
# any number of arguments can have defult values 
def addNo( a, b, c=0):
  sum = a+b+c
   
  return (sum)
```

```
addNo(10,20,30)
addNo(10,20)
```

```
#SyntaxError : non-default argument follows default argument
def addNo( a, b=0, c):
  sum = a+b+c
   
  return (sum)
```

```
#SyntaxError: positional argument follows keyword argument
addNo(b=100,a=30,90)
```

### isinstance method

```
a=10
if (type(a)=='int'): print("yes")
else : print('no')
print(type(a))
#print(isinstance(a,type(10)))
print(isinstance(a,int))
#print(type (10))
```

```
def sumAll(fvar, *otherVariables):
  sum = 0
  sum += fvar
  for i in otherVariables:
    if (isinstance (i , int)) :
      sum += i
  print(sum)
```

```
sumAll(10,20,30)
sumAll(10,20,30,40.44)
sumAll(10,20,30,'sita')
```

```
# add int and float
def sumAll(fvar, *otherVariables):
  sum = 0
  sum += fvar
  for i in otherVariables:
    if (isinstance (i , int)) or (isinstance (i , float)) :
      sum += i
  print(sum)
```

```
sumAll(10,20,30,40.44)
sumAll(10,2.90,30,'sita')
```

## Recursion

```
# calling the method from itself

def factorial (num) : 
  res = 1;
  for i in range (1,num+1):
    res *=i
  return res
```

```
print(factorial(4)) # 4*3*2*1 = 24
print(factorial(5))
```
