## Broadcasting


```
import numpy as np

A = np.arange( 6).reshape(3,2)
print(A.ndim,A.shape)
print(A)
B = np.array([4,8])
print(B.ndim)
print(B,B.shape)
```

```
print(A+B)
#2 new rows are added in B by boardcasting
```

```
C = np.array([4,8,2]).T
print(C.ndim, C.shape, C)
D = C.reshape(3,1)
print(D.ndim, D.shape, D)
```

```
print(A,"\nD: \n" , D," \n A+D : \n", A+D)
#a new column is added by broadcasting
```

## computational efficiency

```
def compute_reciprocal(values) : 
	output = np.empty(len(values))
	for i in range(len(values)):
		output[i] = 1.0/values[i]
	return output
```

```
v1 = np.random.randint(10,100, size=5)
print(v1)
```

```
print(compute_reciprocal(v1))
```

```
%timeit compute_reciprocal(v1)
```

```
%timeit 1/v1
```

```
bigv1 = np.random.randint(1,100, size=100000)
%timeit compute_reciprocal(bigv1) # the simple way in which all languagues does
```

```
%timeit (1.0/bigv1) # the way numpy does
```
