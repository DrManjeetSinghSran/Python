#### Series

```
#pd.Series(data=None, index=None, dtype=None, name=None, copy=False, fastpath=False)

s = pd.Series([1,2,3,4])
s1 = pd.Series(["amit" , 'sumit','sita',1,'hargun'])
#print(s,"\n", s1)
# create own index 
s2 = pd.Series(["amit" , 'sumit','sita','hargun'] , [10,20,30,40])
s3 = pd.Series(  [10,20,30,40],["amit" , 'sumit','sita','hargun'])
print(s3)
# to retrieve value stored at index amit in s3
print(s3['amit'])
print(s2[30])
print(s1[1])
```

```
dict1 = {'mango':'yellow', 'pineapple' : 'green' , 'apple':'red' , 'pomegranate' : 'red'}
#print(dict1)
#print(dict1['apple'])
fruits = pd.Series(dict1)
print(fruits)
# the index of dictionary can be changed in series 
fruits1 = pd.Series(dict1,[ 'pineapple', 'apple', 'pomegranate', 'mango'])
print(fruits1)
print( fruits1[['mango','apple']])
```

```
data = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data, index=['b','c','d','a'])
print (s)
```

```
import pandas as pd
#Create a DataFrame from List of Dicts
data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]
df1 = pd.DataFrame(data)
print(df1)
df2 = pd.DataFrame(data, index=['first', 'second'])
print(df2)
data = [{'a': i, 'b': 2 * i} for i in range(3)]
print (data)
df3 = pd.DataFrame(data)
print(df3)
```

```
#read a list containing student names
students = ['sita','ram','ajmal','altaf','kalam','kapany']
print(students)
data = [{'Rollno': i, 'Name': students[i-1]} for i in range(1, len(students)+1)]
print(data)
dfStudents = pd.DataFrame(data)
print(dfStudents)
#adding new column to dataframe 
dfStudents['Marks'] = 0
print(dfStudents)
dfStudents['Marks'] = dfStudents['Rollno'] +50
print(dfStudents)

marks = {2:90,4:78,1:45,6:67,3:88,5:66}
print(marks)
# add another column named marks along with Rollno and Name in the existing dataframe
 
#dfStudents1 = pd.DataFrame(data.append(marks))
#print(dfStudents1)

dfStudents['markss'] = dfStudents['Rollno'].map(marks) 
print(dfStudents)
```
