## Class Definition

```
# Lovekaran
# make subjects storage flexible
# make space for marks 
# display percentage marks of Student

# create book class (bookcode, book name , authorname, subjectname....)
# each student can have more thank one book

class StudentClass:
  StudentList = None 
  def __init__ (self) :
    self.StudentList = []

  def addStudent(self, name, rollno, subject1,subject2) :
    s = Student( name, rollno, subject1,subject2)
    self.StudentList.append(s)

  def showStudents(self) :
    for  s in self.StudentList : 
      s.display()

class Student: 
    # Constructor 
    def __init__(self, name, rollno, subject1,subject2): 
        self.name = name 
        self.rollno = rollno 
        self.subject1 = subject1 
        self.subject2 = subject2 
 
    # Function to display student details      
    def display(self): 
        print("Name   : ", self.name) 
        print("RollNo : ", self.rollno) 
        print("Subject1 : ", self.subject1) 
        print("Subject2 : ", self.subject2) 
        print("\n") 
class Book:
    BookAccessionNo = 0
    def __init__(self,  title,author,publisher,price):
        Book.BookAccessionNo +=  1
        self.title = title
        self.author = author
        self.publisher = publisher
        self.price = price

    def display (self) : 
      print("Book Accession No  : ", self.BookAccessionNo) 
      print("Title        : ", self.title) 
      print("Author       : ", self.author) 
      print("Publisher    : ", self.publisher) 
      print("Price        : ", self.price) 
      print("\n") 
```

## Object Instantiation

```
# creating Student class & adding students in it
AIML6m = StudentClass()
AIML6m.addStudent('Amit', 11, 'English' , 'Hindi')
AIML6m.addStudent('Robin hood', 12, 'English' , 'Punjabi')
AIML6m.showStudents()
```
