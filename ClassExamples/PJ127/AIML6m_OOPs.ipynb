## Inheritance

Basic Syntax
```
class derived-class(base class):  
    <class-suite>
```
Multiclass (multiple) Inheritance
```
class derive-class(<base class 1>, <base class 2>, ..... <base class n>):  
    <class - suite>  
```
Multilevel Inheritance
```
class class1:  
    <class-suite>   
class class2(class1):  
    <class suite>  
class class3(class2):  
    <class suite> 
``` 

````
class A :
  #constructor 
  def __init__(self) :
    print('A constructor')
    self.x=10
  def setA(self, xx):
    self.x = xx
  def show(self) :
    print("x =" ,self.x)

class B (A):
  #constructor 
  def __init__(self) :
    super().__init__()    # calling base class(A) constructor
    print('B constructor')
    self.y=5
  def showB(self): 
    print("x =" ,self.x)
    print("y =" ,self.y)
````

````
a1 = A()
a1.show()
a1.setA(100)
a1.show()
````

````
b1 = B()
b1.show()
b1.setA(40)
#b1.show()
b1.showB()
````

#### Method Overriding

````
class A :
  #constructor 
  def __init__(self) :
    print('A constructor')
    self.x=10
  def setA(self, xx):
    self.x = xx
  def show(self) :
    print("A show")
    print("x =" ,self.x)

class B (A):    # B is derived class A is base class
  #constructor 
  def __init__(self) :
    super().__init__()    # calling base class(A) constructor
    print('B constructor')
    self.y=5
  def show(self):     # override the base class method show(self)
    super().show()   # calls base class method using super()
    print("B show")
    print("x =" ,self.x)
    print("y =" ,self.y)
# overriding - if same methodname or variable name is used in derived class 
# in simple word the derived class variable has hidden the basae class variable
# to call the hidden method use super() keyword
````

````
b1 = B()
b1.show()
````

#### Destructor in inheritance

````
class A :
  #constructor 
  def __init__(self) :
    print('A constructor')
   
  def __del__(self ) : 
    print('A destructor ')

class B (A):    # B is derived class A is base class
  #constructor 
  def __init__(self) :
    super().__init__()    # calling base class(A) constructor
    print('B constructor')

  def __del__(self ) : 
    super().__del__()
    print('B destructor ')
````

### Data hiding (public, private, protected)

image.png

````
# by default all variables are public 

class A :
  
  iPub = None    # public data member
  _iPro = None   # protected data member 
  __iPri = None  # private data member 

  def __init__ (self, iPub=0,iPro=0,iPri=0):

    self.iPub = iPub
    self._iPro= iPro     #   protected attribute 
    self.__iPri= iPri     #   private attribute 

  def showA(self) :
    print ( "iPub : ",self.iPub , ", iPro : ",self._iPro, ", iPri : ", self.__iPri)

  def __str__(self) :
    return "iPub : {0}  , iPro : {1} , iPri : {2} ".format(  self.iPub,  self._iPro, self.__iPri ) 
````

````
#a1 = A()
a1 = A(10,20,30)
a1.showA()
print(a1)
````

````
a1.iPub = 100
a1.showA()
a1._iPro = 70
a1.showA()
# a1.iPri = 20  cannot access private like this
a1.__iPri = 20   # private cannot be accessed using __
a1._A__iPri = 20   # private can be accessed using _className__privateVariableName
a1.showA()
print(a1)
````

**Protected Access Modifier**

 * The members of a class that are declared protected are only accessible to a class derived from it. Data members of a class are declared protected by adding a single underscore ‘_’ symbol before the data member of that class.

**Private Access Modifier**

* The members of a class that are declared private are accessible within the class only, private access modifier is the most secure access modifier. Data members of a class are declared private by adding a double underscore ‘__’ symbol before the data member of that class.

* Every member with double underscore will be changed to _object._class__variable. If so required, it can still be accessed from outside the class, but the practice should be refrained. a1.**_class__iPri**

### Public, Private, Protected Methods 

````
class A : 
  iPub = None    # public data member
  _iPro = None   # protected data member 
  __iPri = None  # private data member 

  def __init__ (self, iPub=0,iPro=0,iPri=0):
    self.iPub = iPub
    self._iPro= iPro     #   protected attribute 
    self.__iPri= iPri     #   private attribute 

  def showiPub(self ) : 
    print( "iPib : " , self.iPub)

  def _showiPro(self ) : 
    print( "iPro : " , self._iPro)  

  def __showiPri(self ) : 
    print( "iPri : " , self.__iPri) 
````
