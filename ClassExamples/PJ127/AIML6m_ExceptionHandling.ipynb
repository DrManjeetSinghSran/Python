## Raise Exception

```
#accept age from user and display whether he is able to vote

age = eval(input('Enter your age   '))
try :

  if (age<1) : 
    raise ValueError(' Age cannot be less than zero ')
  if (age>18) : print('U can vote')
  else : print('Cannot vote')

except ValueError: 
  print(' Error Occured ')
except :
  print(' Any other error ')
```

```
#accept age from user and display whether he is able to vote
# display the message as Error object is initialised

age = eval(input('Enter your age(in years)  : '))
try :

  if (age<1) : 
    raise ValueError(' Age cannot be less than zero ')
  if (age>150) :
    raise ValueError(' Age cannot be greater than 150 years ')
  if (age>18) : print('U can vote')
  else : print('Cannot vote')

except ValueError as a: 
  print(' Age Value Error: ',a)
except :
  print(' Any other error ')
```

```
# exception in methods
def getAge():
  age = eval(input('Enter your age(in years)  : '))
  if (age<1) : 
    raise ValueError(' Age cannot be less than zero ')
  if (age>150) :
    raise ValueError(' Age cannot be greater than 150 years ')
  if (age>18) : print('U can vote')
  else : print('Cannot vote')
  return age
```

```
#calling the method and handling the exception
try:
  x= getAge(  )
  print (x )
except ValueError as arg : 
  print("Age Error : ",arg ) 
```

```
# multiple method calling

def getStudentDetails():
  Name = input ('Enter Student name : ')
  Age = getAge()
  print("{0} is {1} years old".format(Name, Age))
```

```
try :
  getStudentDetails()
except ValueError as arg : 
  print("Age Error : ",arg )
```

## User defined Exception

```
#create the user defined exception class in .py file 
class NegativeNumberException(RuntimeError):
  def __init(self,age):
    super.__init__()
    self.age = age
```

```
a = eval(input (' Enter Age(in years) : '))
try:
  
  if a<0 : 
    raise NegativeNumberException('age cannot be less than zero ')

except NegativeNumberException as n :
  print('Number exception : ',n)  
```

## assert Statment

Assertions are statements that assert or state a fact confidently in your program.

assert Expression[, Arguments]

```
# calculate average of student marks 
marks =[78,87,65,67]
#marks=[]
print(marks)
#assert len(marks) != 0
assert len(marks) != 0 , ' no marks '
avg = sum(marks)/len(marks)
print(avg)
```

```
#define method to calculate average of student marks
def calAverage ( marks):
  assert len(marks) != 0 , "  no marks provided  "
  avg = sum(marks)/len(marks)
  return (avg)  
```

```
marks =[78,87,65,67]
print(calAverage(marks) )
m=[]
print( calAverage(m))
```
